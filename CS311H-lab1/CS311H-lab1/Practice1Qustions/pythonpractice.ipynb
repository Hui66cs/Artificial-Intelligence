## Python Practice (0.5 points)
### Snake: Simulate a snake game and print the game results.

You are given a map that ‘x’ represents a rock, ‘-’represents a space, ‘#’represents the body of snake. ‘@’represent the head of the snake and a sequence of actions that ‘0,1,2,3’represent to move to up/down/left/right correspondingly for one step.
A greedy snake starts in the map state and moves one step per unit of time according to the sequence of actions until all actions complete or fail. It will fail when the head and the stone overlap, the head goes beyond the boundary, or the head overlaps the body. 

#### Input
A matrix with type char (the map). 
A sequence with type int (the motions). 

#### Output
The result of the game:
If it failed, output the running time of the game.
If it didn’t fail, output the final position of the head (in the form “%d, %d”).

```python
"""
Example:
input:
map:
---------
------x--
-x-------
---@-----
---##----
------x--
--x----x-
-x-------
---------
action:
0 0 3 3 0 3 3 1 1 1 1 1 3 1 1 2 2 2 2 2

output:
7 3
"""
```

```python
# add your code here
def solve(game_map,actions):

    def check(x,y,body):
        if x < 0 or x >= len(game_map) or y < 0 or y >= len(game_map[0]):
            return False
        if game_map[x][y] == 'x':
            return False
        return True     

    def find_body(x,y,s,game_map,body):  #根据上一节身体（一开始是头）找下一节身体
        if check(x+1,y,body) and game_map[x+1][y] == '#' and (x+1,y) not in s:
            return [x+1,y]
        elif check(x-1,y,body) and game_map[x-1][y] == '#' and (x-1,y) not in s:
            return [x-1,y]
        elif check(x,y+1,body) and game_map[x][y+1] == '#' and (x,y+1) not in s:
            return [x,y+1]
        elif check(x,y-1,body) and game_map[x][y-1] == '#' and (x,y-1) not in s:
            return [x,y-1]
        else:
            return None

    body = []
    for i in range (0,len(game_map)):
        for j in range (0,len(game_map[0])):
            if game_map[i][j] == '@':
                head_x = i
                head_y = j
                break
    temp_x = head_x
    temp_y = head_y
    s=set()
    while find_body(head_x, head_y,s, game_map,body) != None:
        [head_x,head_y] = find_body(head_x, head_y, s, game_map,body)
        s.add((head_x,head_y))
        body.append([head_x,head_y])

    time = 0
    head_x = temp_x
    head_y = temp_y
    IsFail = False

    for i in range (0, len(actions)):
        time+=1
        if actions[i] == 0:
            head_x -= 1
        elif actions[i] == 1:
            head_x += 1
        elif actions[i] == 2:
            head_y -= 1
        elif actions[i] == 3:
            head_y += 1
        if not check(head_x, head_y,body):
            print(time)
            IsFail=True
            break
        else:
            for j in range (len(body)-1, 0, -1):
                body[j][0] = body[j-1][0]
                body[j][1] = body[j-1][1]
            body[0][0] = temp_x
            body[0][1] = temp_y
            if [head_x, head_y] in body:
                print(time)
                IsFail=True
                break
            temp_x = head_x
            temp_y = head_y
    if not IsFail:
        print("%d,%d" % (head_x, head_y))
```
